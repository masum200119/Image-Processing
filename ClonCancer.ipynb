```python
import os
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms
from torch.optim import AdamW
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import (accuracy_score, classification_report, 
                           ConfusionMatrixDisplay, confusion_matrix, roc_curve, auc,
                           precision_recall_curve, average_precision_score, f1_score,
                           precision_score, recall_score)
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import StratifiedKFold
import numpy as np
from itertools import cycle
from efficientnet_pytorch import EfficientNet
import random
import seaborn as sns
from collections import defaultdict

# Configuration
CONFIG = {
    "dataset_path": r"D:\New folder\Downloads\ColonCancerCT-2025 A Dataset of Abdominal CT Scans",
    "image_size": 224,
    "batch_size": 16,
    "num_epochs": 20,
    "learning_rate": 5e-5,
    "weight_decay": 0.01,
    "dropout_rate": 0.2,
    "seed": 42
}

# Set random seeds for reproducibility
def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(CONFIG["seed"])

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Data preparation
def prepare_data():
    # Common transform without augmentation
    base_transform = transforms.Compose([ 
        transforms.Resize((CONFIG["image_size"], CONFIG["image_size"])),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    # Augmentation for training
    train_transform = transforms.Compose([ 
        transforms.Resize((CONFIG["image_size"], CONFIG["image_size"])),
        transforms.RandomHorizontalFlip(),
        transforms.RandomVerticalFlip(),
        transforms.RandomRotation(30),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
        transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
        transforms.RandomResizedCrop(CONFIG["image_size"], scale=(0.8, 1.0)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        transforms.RandomErasing(p=0.1, scale=(0.02, 0.1), ratio=(0.3, 3.3))
    ])
    
    # Load datasets
    train_dataset = datasets.ImageFolder(os.path.join(CONFIG["dataset_path"], "train"), transform=train_transform)
    val_dataset = datasets.ImageFolder(os.path.join(CONFIG["dataset_path"], "val"), transform=base_transform)
    test_dataset = datasets.ImageFolder(os.path.join(CONFIG["dataset_path"], "test"), transform=base_transform)
    
    print("\nClass distribution:")
    print(f"Train: {np.bincount([y for _, y in train_dataset])}")
    print(f"Val: {np.bincount([y for _, y in val_dataset])}")
    print(f"Test: {np.bincount([y for _, y in test_dataset])}")
    
    return (
        DataLoader(train_dataset, batch_size=CONFIG["batch_size"], shuffle=True, num_workers=2, pin_memory=True),
        DataLoader(val_dataset, batch_size=CONFIG["batch_size"], shuffle=False, num_workers=2, pin_memory=True),
        DataLoader(test_dataset, batch_size=CONFIG["batch_size"], shuffle=False, num_workers=2, pin_memory=True),
        train_dataset.classes
    )

# Model setup
def setup_model(num_classes):
    model = EfficientNet.from_pretrained('efficientnet-b0')
    num_ftrs = model._fc.in_features
    model._fc = nn.Sequential(
        nn.Dropout(p=CONFIG["dropout_rate"]),
        nn.Linear(num_ftrs, num_classes)
    )
    for param in model.parameters():
        param.requires_grad = False
    for param in model._blocks[-5:].parameters():
        param.requires_grad = True
    for param in model._conv_head.parameters():
        param.requires_grad = True
    for param in model._bn1.parameters():
        param.requires_grad = True
    for param in model._fc.parameters():
        param.requires_grad = True
    
    return model.to(device)

# Training loop
def train_model(model, train_loader, val_loader):
    optimizer = AdamW(
        filter(lambda p: p.requires_grad, model.parameters()), 
        lr=CONFIG["learning_rate"], 
        weight_decay=CONFIG["weight_decay"]
    )
    criterion = nn.CrossEntropyLoss()
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='max', factor=0.1, patience=3, verbose=True
    )
    
    history = {
        'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': [],
        'train_precision': [], 'val_precision': [], 'train_recall': [], 'val_recall': [],
        'train_f1': [], 'val_f1': [], 'learning_rates': []
    }
    
    for epoch in range(CONFIG["num_epochs"]):
        model.train()
        train_loss, train_preds, train_labels = 0.0, [], []
        
        for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{CONFIG['num_epochs']} - Training"):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            train_preds.extend(torch.argmax(outputs, dim=1).cpu().numpy())
            train_labels.extend(labels.cpu().numpy())
        
        train_loss /= len(train_loader)
        train_acc = accuracy_score(train_labels, train_preds)
        train_precision = precision_score(train_labels, train_preds, average='weighted', zero_division=0)
        train_recall = recall_score(train_labels, train_preds, average='weighted', zero_division=0)
        train_f1 = f1_score(train_labels, train_preds, average='weighted', zero_division=0)
        
        # Validation
        val_loss, val_preds, val_labels = evaluate_model(model, val_loader, criterion)
        val_acc = accuracy_score(val_labels, val_preds)
        val_precision = precision_score(val_labels, val_preds, average='weighted', zero_division=0)
        val_recall = recall_score(val_labels, val_preds, average='weighted', zero_division=0)
        val_f1 = f1_score(val_labels, val_preds, average='weighted', zero_division=0)
        
        # Store metrics
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        history['train_precision'].append(train_precision)
        history['val_precision'].append(val_precision)
        history['train_recall'].append(train_recall)
        history['val_recall'].append(val_recall)
        history['train_f1'].append(train_f1)
        history['val_f1'].append(val_f1)
        history['learning_rates'].append(optimizer.param_groups[0]['lr'])
        
        scheduler.step(val_acc)
        
        print(f"Epoch {epoch+1}: Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")
        print(f"         Train F1: {train_f1:.4f} | Val F1: {val_f1:.4f} | LR: {optimizer.param_groups[0]['lr']:.6f}")
    
    return history, val_preds, val_labels

def evaluate_model(model, loader, criterion):
    model.eval()
    total_loss, preds, labels = 0.0, [], []
    with torch.no_grad():
        for inputs, labels_batch in tqdm(loader, desc="Evaluating"):
            inputs, labels_batch = inputs.to(device), labels_batch.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels_batch)
            total_loss += loss.item()
            preds.extend(torch.argmax(outputs, dim=1).cpu().numpy())
            labels.extend(labels_batch.cpu().numpy())
    
    return total_loss / len(loader), preds, labels

def get_predictions_with_probabilities(model, loader):
    model.eval()
    all_preds = []
    all_labels = []
    all_probs = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(loader, desc="Getting predictions with probabilities"):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            probs = torch.softmax(outputs, dim=1)
            
            all_preds.extend(torch.argmax(outputs, dim=1).cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())
    
    return np.array(all_preds), np.array(all_labels), np.array(all_probs)

def plot_comprehensive_results(history, class_names, val_preds, val_labels, test_preds=None, test_labels=None, test_probs=None):
    # Create a comprehensive figure with multiple subplots
    fig = plt.figure(figsize=(25, 30))
    
    # 1. Training History - Loss and Accuracy
    plt.subplot(5, 4, 1)
    plt.plot(history['train_loss'], label='Training Loss', color='blue', linewidth=2)
    plt.plot(history['val_loss'], label='Validation Loss', color='red', linewidth=2)
    plt.title('Training and Validation Loss', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(5, 4, 2)
    plt.plot(history['train_acc'], label='Training Accuracy', color='blue', linewidth=2)
    plt.plot(history['val_acc'], label='Validation Accuracy', color='red', linewidth=2)
    plt.title('Training and Validation Accuracy', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 2. Precision, Recall, and F1-Score
    plt.subplot(5, 4, 3)
    plt.plot(history['train_precision'], label='Training Precision', color='blue', linewidth=2)
    plt.plot(history['val_precision'], label='Validation Precision', color='red', linewidth=2)
    plt.title('Training and Validation Precision', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('Precision')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(5, 4, 4)
    plt.plot(history['train_recall'], label='Training Recall', color='blue', linewidth=2)
    plt.plot(history['val_recall'], label='Validation Recall', color='red', linewidth=2)
    plt.title('Training and Validation Recall', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('Recall')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(5, 4, 5)
    plt.plot(history['train_f1'], label='Training F1-Score', color='blue', linewidth=2)
    plt.plot(history['val_f1'], label='Validation F1-Score', color='red', linewidth=2)
    plt.title('Training and Validation F1-Score', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('F1-Score')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 3. Learning Rate Schedule
    plt.subplot(5, 4, 6)
    plt.plot(history['learning_rates'], color='green', linewidth=2)
    plt.title('Learning Rate Schedule', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.yscale('log')
    plt.grid(True, alpha=0.3)
    
    # 4. Confusion Matrix for Validation
    plt.subplot(5, 4, 7)
    cm_val = confusion_matrix(val_labels, val_preds)
    sns.heatmap(cm_val, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Validation Confusion Matrix', fontsize=14)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    
    # 5. Confusion Matrix for Test (if available)
    if test_preds is not None and test_labels is not None:
        plt.subplot(5, 4, 8)
        cm_test = confusion_matrix(test_labels, test_preds)
        sns.heatmap(cm_test, annot=True, fmt='d', cmap='Greens', 
                    xticklabels=class_names, yticklabels=class_names)
        plt.title('Test Confusion Matrix', fontsize=14)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
    
    # 6. ROC Curves (if test probabilities available)
    if test_probs is not None and test_labels is not None:
        plt.subplot(5, 4, 9)
        n_classes = len(class_names)
        test_labels_bin = label_binarize(test_labels, classes=range(n_classes))
        
        if n_classes == 2:
            fpr, tpr, _ = roc_curve(test_labels_bin, test_probs[:, 1])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})', linewidth=2)
        else:
            colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'red', 'green', 'purple'])
            for i, color in zip(range(n_classes), colors):
                fpr, tpr, _ = roc_curve(test_labels_bin[:, i], test_probs[:, i])
                roc_auc = auc(fpr, tpr)
                plt.plot(fpr, tpr, color=color, lw=2,
                        label=f'ROC curve of class {class_names[i]} (AUC = {roc_auc:.2f})')
        
        plt.plot([0, 1], [0, 1], 'k--', lw=2)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curves', fontsize=14)
        plt.legend(loc="lower right")
        plt.grid(True, alpha=0.3)
    
    # 7. Precision-Recall Curves
    if test_probs is not None and test_labels is not None:
        plt.subplot(5, 4, 10)
        n_classes = len(class_names)
        test_labels_bin = label_binarize(test_labels, classes=range(n_classes))
        
        if n_classes == 2:
            precision, recall, _ = precision_recall_curve(test_labels_bin, test_probs[:, 1])
            avg_precision = average_precision_score(test_labels_bin, test_probs[:, 1])
            plt.plot(recall, precision, label=f'PR curve (AP = {avg_precision:.2f})', linewidth=2)
        else:
            colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'red', 'green', 'purple'])
            for i, color in zip(range(n_classes), colors):
                precision, recall, _ = precision_recall_curve(test_labels_bin[:, i], test_probs[:, i])
                avg_precision = average_precision_score(test_labels_bin[:, i], test_probs[:, i])
                plt.plot(recall, precision, color=color, lw=2,
                        label=f'PR curve of class {class_names[i]} (AP = {avg_precision:.2f})')
        
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Precision-Recall Curves', fontsize=14)
        plt.legend()
        plt.grid(True, alpha=0.3)
    
    # 8. Class-wise Performance Metrics
    if test_preds is not None and test_labels is not None:
        plt.subplot(5, 4, 11)
        report = classification_report(test_labels, test_preds, target_names=class_names, output_dict=True)
        
        metrics = ['precision', 'recall', 'f1-score']
        class_metrics = {metric: [] for metric in metrics}
        
        for class_name in class_names:
            for metric in metrics:
                class_metrics[metric].append(report[class_name][metric])
        
        x = np.arange(len(class_names))
        width = 0.25
        
        for i, metric in enumerate(metrics):
            plt.bar(x + i*width, class_metrics[metric], width, label=metric.capitalize())
        
        plt.xlabel('Classes')
        plt.ylabel('Score')
        plt.title('Class-wise Performance Metrics', fontsize=14)
        plt.xticks(x + width, class_names, rotation=45)
        plt.legend()
        plt.grid(True, alpha=0.3)
    
    # 9. Training Progress Summary
    plt.subplot(5, 4, 12)
    epochs = range(1, len(history['train_loss']) + 1)
    plt.plot(epochs, history['train_loss'], 'b-', label='Training Loss', linewidth=2)
    plt.plot(epochs, history['val_loss'], 'r-', label='Validation Loss', linewidth=2)
    plt.plot(epochs, [1-acc for acc in history['train_acc']], 'b--', label='Training Error', linewidth=2)
    plt.plot(epochs, [1-acc for acc in history['val_acc']], 'r--', label='Validation Error', linewidth=2)
    plt.title('Training Progress: Loss and Error', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('Loss / Error')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 10. Validation Accuracy vs Loss
    plt.subplot(5, 4, 13)
    plt.scatter(history['val_loss'], history['val_acc'], c=range(len(history['val_loss'])), 
                cmap='viridis', s=50)
    plt.xlabel('Validation Loss')
    plt.ylabel('Validation Accuracy')
    plt.title('Validation Accuracy vs Loss', fontsize=14)
    plt.colorbar(label='Epoch')
    plt.grid(True, alpha=0.3)
    
    # 11. Learning Rate vs Validation Accuracy
    plt.subplot(5, 4, 14)
    plt.semilogx(history['learning_rates'], history['val_acc'], 'o-', linewidth=2, markersize=4)
    plt.xlabel('Learning Rate')
    plt.ylabel('Validation Accuracy')
    plt.title('Learning Rate vs Validation Accuracy', fontsize=14)
    plt.grid(True, alpha=0.3)
    
    # 12. Training Speed (Loss Reduction Rate)
    plt.subplot(5, 4, 15)
    loss_reduction = []
    for i in range(1, len(history['train_loss'])):
        reduction = history['train_loss'][i-1] - history['train_loss'][i]
        loss_reduction.append(reduction)
    
    plt.plot(range(2, len(history['train_loss']) + 1), loss_reduction, 'g-', linewidth=2)
    plt.xlabel('Epoch')
    plt.ylabel('Loss Reduction')
    plt.title('Training Loss Reduction Rate', fontsize=14)
    plt.axhline(y=0, color='r', linestyle='--', alpha=0.5)
    plt.grid(True, alpha=0.3)
    
    # 13. Overfitting Analysis
    plt.subplot(5, 4, 16)
    overfitting_gap = [train_acc - val_acc for train_acc, val_acc in zip(history['train_acc'], history['val_acc'])]
    plt.plot(range(1, len(overfitting_gap) + 1), overfitting_gap, 'purple', linewidth=2)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy Gap (Train - Val)')
    plt.title('Overfitting Analysis', fontsize=14)
    plt.axhline(y=0, color='r', linestyle='--', alpha=0.5)
    plt.grid(True, alpha=0.3)
    
    # 14. Metric Correlation Matrix
    plt.subplot(5, 4, 17)
    metrics_df = {
        'Train Loss': history['train_loss'],
        'Val Loss': history['val_loss'],
        'Train Acc': history['train_acc'],
        'Val Acc': history['val_acc'],
        'Train F1': history['train_f1'],
        'Val F1': history['val_f1']
    }
    correlation_matrix = np.corrcoef([metrics_df[key] for key in metrics_df.keys()])
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
                xticklabels=list(metrics_df.keys()), yticklabels=list(metrics_df.keys()))
    plt.title('Metric Correlation Matrix', fontsize=14)
    
    # 15. Per-class Accuracy (if available)
    if test_preds is not None and test_labels is not None:
        plt.subplot(5, 4, 18)
        per_class_acc = []
        for i, class_name in enumerate(class_names):
            class_mask = np.array(test_labels) == i
            if np.sum(class_mask) > 0:
                class_acc = accuracy_score(np.array(test_labels)[class_mask], 
                                         np.array(test_preds)[class_mask])
                per_class_acc.append(class_acc)
            else:
                per_class_acc.append(0)
        
        plt.bar(range(len(class_names)), per_class_acc, color='skyblue')
        plt.xlabel('Classes')
        plt.ylabel('Accuracy')
        plt.title('Per-class Accuracy', fontsize=14)
        plt.xticks(range(len(class_names)), class_names, rotation=45)
        plt.grid(True, alpha=0.3)
    
    # 16. Training Stability - Rolling Standard Deviation
    plt.subplot(5, 4, 19)
    window_size = 3
    rolling_train_std = []
    rolling_val_std = []
    
    for i in range(len(history['train_loss'])):
        start_idx = max(0, i - window_size + 1)
        end_idx = i + 1
        train_window = history['train_loss'][start_idx:end_idx]
        val_window = history['val_loss'][start_idx:end_idx]
        
        rolling_train_std.append(np.std(train_window))
        rolling_val_std.append(np.std(val_window))
    
    plt.plot(range(1, len(rolling_train_std) + 1), rolling_train_std, 'b-', 
             label='Train Loss Std', linewidth=2)
    plt.plot(range(1, len(rolling_val_std) + 1), rolling_val_std, 'r-', 
             label='Val Loss Std', linewidth=2)
    plt.xlabel('Epoch')
    plt.ylabel('Loss Standard Deviation')
    plt.title('Training Stability (Rolling Std)', fontsize=14)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 17. Epoch-wise Performance Summary
    plt.subplot(5, 4, 20)
    performance_score = [(acc + f1) / 2 for acc, f1 in zip(history['val_acc'], history['val_f1'])]
    plt.plot(range(1, len(performance_score) + 1), performance_score, 'orange', linewidth=3)
    plt.xlabel('Epoch')
    plt.ylabel('Combined Performance Score')
    plt.title('Overall Performance Trend', fontsize=14)
    plt.grid(True, alpha=0.3)
    
    # Mark best epoch
    best_epoch = np.argmax(performance_score) + 1
    plt.axvline(x=best_epoch, color='red', linestyle='--', alpha=0.7)
    plt.text(best_epoch, max(performance_score), f'Best: {best_epoch}', 
             rotation=90, verticalalignment='bottom')
    
    plt.tight_layout()
    plt.savefig('comprehensive_training_results.png', dpi=300, bbox_inches='tight')
    plt.show()

# Main execution
if __name__ == "__main__":
    # Prepare data
    train_loader, val_loader, test_loader, class_names = prepare_data()
    
    # Setup model
    model = setup_model(len(class_names))
    print(f"\nModel has {sum(p.numel() for p in model.parameters() if p.requires_grad)} trainable parameters")
    
    # Train model
    print("\nStarting training...")
    history, val_preds, val_labels = train_model(model, train_loader, val_loader)
    
    # Evaluate on test set
    print("\nEvaluating on test set...")
    test_preds, test_labels, test_probs = get_predictions_with_probabilities(model, test_loader)
    
    # Plot comprehensive results
    print("\nGenerating visualizations...")
    plot_comprehensive_results(history, class_names, val_preds, val_labels, test_preds, test_labels, test_probs)
    
    # Save model
    torch.save(model.state_dict(), 'final_model.pth')
    print("\nTraining completed and model saved!")
```

```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from torchvision import datasets, transforms
from torch.optim import AdamW
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import (accuracy_score, classification_report, 
                           ConfusionMatrixDisplay, confusion_matrix, roc_curve, auc,
                           precision_recall_curve, average_precision_score, f1_score,
                           precision_score, recall_score)
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import StratifiedKFold
import numpy as np
from itertools import cycle
import timm
import random
import seaborn as sns
from collections import defaultdict, Counter
import warnings
warnings.filterwarnings('ignore')

# Enhanced Configuration for Large Dataset
CONFIG = {
    "dataset_path": r"D:\New folder\Downloads\ColonCancerCT-2025 A Dataset of Abdominal CT Scans",
    "image_size": 224,  # Reduced for RTX 3050 Ti (4GB VRAM)
    "batch_size": 16,   # Reduced for memory constraints
    "num_epochs": 30,   # Reduced for faster training
    "learning_rate": 1e-4,
    "weight_decay": 0.01,
    "dropout_rate": 0.3,
    "seed": 42,
    "model_name": "efficientnet_b0",  # Lighter model for 4GB GPU
    "use_focal_loss": True,
    "label_smoothing": 0.1,
    "mixup_alpha": 0.2,  # Reduced for stability
    "cutmix_alpha": 0.2,
    "early_stopping_patience": 8,
    "cosine_restarts": False,  # Simplified for stability
    "gradient_clip_val": 1.0,
    "accumulate_grad_batches": 4,  # Increased for effective batch size
    "use_sam_optimizer": False,  # Disabled for memory efficiency
    "test_time_augmentation": True
}

# Set random seeds for reproducibility
def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(CONFIG["seed"])

# Device setup with mixed precision
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB")

# Enhanced Data Augmentation
def get_transforms():
    # Balanced augmentation for training
    train_transform = transforms.Compose([
        transforms.Resize((CONFIG["image_size"], CONFIG["image_size"])),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(degrees=15),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
        transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.9, 1.1)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        transforms.RandomErasing(p=0.1, scale=(0.02, 0.10), ratio=(0.3, 3.3))
    ])
    
    # Standard transforms for validation/test
    val_transform = transforms.Compose([
        transforms.Resize((CONFIG["image_size"], CONFIG["image_size"])),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    # Test Time Augmentation transforms
    tta_transforms = [
        transforms.Compose([
            transforms.Resize((CONFIG["image_size"], CONFIG["image_size"])),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ]),
        transforms.Compose([
            transforms.Resize((CONFIG["image_size"], CONFIG["image_size"])),
            transforms.RandomHorizontalFlip(p=1.0),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    ]
    
    return train_transform, val_transform, tta_transforms

# Mixup/Cutmix Implementation
def mixup_data(x, y, alpha=1.0):
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1
    
    batch_size = x.size(0)
    index = torch.randperm(batch_size).to(x.device)
    
    mixed_x = lam * x + (1 - lam) * x[index, :]
    y_a, y_b = y, y[index]
    return mixed_x, y_a, y_b, lam

def rand_bbox(size, lam):
    W = size[2]
    H = size[3]
    cut_rat = np.sqrt(1. - lam)
    cut_w = int(W * cut_rat)
    cut_h = int(H * cut_rat)
    
    cx = np.random.randint(W)
    cy = np.random.randint(H)
    
    bbx1 = np.clip(cx - cut_w // 2, 0, W)
    bby1 = np.clip(cy - cut_h // 2, 0, H)
    bbx2 = np.clip(cx + cut_w // 2, 0, W)
    bby2 = np.clip(cy + cut_h // 2, 0, H)
    
    return bbx1, bby1, bbx2, bby2

def cutmix_data(x, y, alpha=1.0):
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1
    
    batch_size = x.size(0)
    index = torch.randperm(batch_size).to(x.device)
    
    bbx1, bby1, bbx2, bby2 = rand_bbox(x.size(), lam)
    x[:, :, bbx1:bbx2, bby1:bby2] = x[index, :, bbx1:bbx2, bby1:bby2]
    
    lam = 1 - ((bbx2 - bbx1) * (bby2 - bby1) / (x.size()[-1] * x.size()[-2]))
    y_a, y_b = y, y[index]
    return x, y_a, y_b, lam

# Enhanced Data Preparation
def prepare_data():
    train_transform, val_transform, tta_transforms = get_transforms()
    
    # Load datasets
    train_dataset = datasets.ImageFolder(os.path.join(CONFIG["dataset_path"], "train"), transform=train_transform)
    val_dataset = datasets.ImageFolder(os.path.join(CONFIG["dataset_path"], "val"), transform=val_transform)
    test_dataset = datasets.ImageFolder(os.path.join(CONFIG["dataset_path"], "test"), transform=val_transform)
    
    # Calculate class weights for balanced training
    class_counts = Counter([y for _, y in train_dataset])
    total_samples = len(train_dataset)
    class_weights = {i: total_samples / (len(class_counts) * count) for i, count in class_counts.items()}
    sample_weights = [class_weights[y] for _, y in train_dataset]
    
    # Create weighted sampler
    sampler = WeightedRandomSampler(sample_weights, len(sample_weights), replacement=True)
    
    print(f"\nDataset Information:")
    print(f"Train samples: {len(train_dataset):,}")
    print(f"Val samples: {len(val_dataset):,}")
    print(f"Test samples: {len(test_dataset):,}")
    print(f"Classes: {train_dataset.classes}")
    print(f"Class distribution (train): {dict(class_counts)}")
    print(f"Class weights: {class_weights}")
    
    # Ensure batch size doesn't cause single-sample batches
    batch_size = min(CONFIG["batch_size"], len(train_dataset))
    
    train_loader = DataLoader(train_dataset, batch_size=batch_size, 
                             sampler=sampler, num_workers=2, pin_memory=True, drop_last=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, 
                           shuffle=False, num_workers=2, pin_memory=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, 
                            shuffle=False, num_workers=2, pin_memory=True)
    
    return train_loader, val_loader, test_loader, train_dataset.classes, class_weights

# Simplified Model with Attention
class AttentionBlock(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        self.attention = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(in_channels, in_channels // 8, 1),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels // 8, in_channels, 1),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        att = self.attention(x)
        return x * att

class EnhancedClassifier(nn.Module):
    def __init__(self, num_classes, model_name="efficientnet_b0"):
        super().__init__()
        self.backbone = timm.create_model(model_name, pretrained=True, num_classes=0)
        
        # Get feature dimension
        with torch.no_grad():
            dummy_input = torch.randn(1, 3, CONFIG["image_size"], CONFIG["image_size"])
            features = self.backbone(dummy_input)
            feature_dim = features.shape[1]
        
        self.attention = AttentionBlock(feature_dim)
        self.global_pool = nn.AdaptiveAvgPool2d(1)
        
        # Enhanced classifier head
        self.classifier = nn.Sequential(
            nn.Dropout(CONFIG["dropout_rate"]),
            nn.Linear(feature_dim, feature_dim // 2),
            nn.ReLU(inplace=True),
            nn.LayerNorm(feature_dim // 2),  # Changed from BatchNorm to LayerNorm
            nn.Dropout(CONFIG["dropout_rate"] / 2),
            nn.Linear(feature_dim // 2, num_classes)
        )
        
    def forward(self, x):
        features = self.backbone(x)
        
        # Reshape for attention if needed
        if len(features.shape) == 2:
            features = features.unsqueeze(-1).unsqueeze(-1)
        
        features = self.attention(features)
        features = self.global_pool(features).flatten(1)
        
        return self.classifier(features)

# Focal Loss Implementation
class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2, num_classes=4):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.num_classes = num_classes
        
    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * ce_loss
        return focal_loss.mean()

# Enhanced Training Function
def train_model(model, train_loader, val_loader, class_weights):
    # Setup loss function
    if CONFIG["use_focal_loss"]:
        criterion = FocalLoss(num_classes=len(class_weights))
    else:
        criterion = nn.CrossEntropyLoss(label_smoothing=CONFIG["label_smoothing"])
    
    # Setup optimizer
    optimizer = AdamW(model.parameters(), lr=CONFIG["learning_rate"], 
                     weight_decay=CONFIG["weight_decay"])
    
    # Enhanced scheduler
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='max', factor=0.5, patience=5, verbose=True
    )
    
    # Mixed precision training
    scaler = torch.cuda.amp.GradScaler()
    
    # Training history
    history = {
        'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': [],
        'train_precision': [], 'val_precision': [], 'train_recall': [], 'val_recall': [],
        'train_f1': [], 'val_f1': [], 'learning_rates': []
    }
    
    best_val_acc = 0.0
    patience_counter = 0
    
    for epoch in range(CONFIG["num_epochs"]):
        # Training phase
        model.train()
        train_loss = 0.0
        train_preds = []
        train_labels = []
        
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{CONFIG['num_epochs']} - Training")
        
        for batch_idx, (inputs, labels) in enumerate(pbar):
            inputs, labels = inputs.to(device), labels.to(device)
            
            # Skip batches with only one sample
            if inputs.size(0) == 1:
                continue
                
            # Apply Mixup/CutMix occasionally
            if random.random() < 0.3:
                if random.random() < 0.5:
                    inputs, labels_a, labels_b, lam = mixup_data(inputs, labels, CONFIG["mixup_alpha"])
                else:
                    inputs, labels_a, labels_b, lam = cutmix_data(inputs, labels, CONFIG["cutmix_alpha"])
                
                # Mixed precision forward pass
                with torch.cuda.amp.autocast():
                    outputs = model(inputs)
                    loss = lam * criterion(outputs, labels_a) + (1 - lam) * criterion(outputs, labels_b)
                    loss = loss / CONFIG["accumulate_grad_batches"]
            else:
                # Standard forward pass
                with torch.cuda.amp.autocast():
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                    loss = loss / CONFIG["accumulate_grad_batches"]
            
            # Backward pass
            scaler.scale(loss).backward()
            
            if (batch_idx + 1) % CONFIG["accumulate_grad_batches"] == 0:
                # Gradient clipping
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), CONFIG["gradient_clip_val"])
                
                scaler.step(optimizer)
                scaler.update()
                optimizer.zero_grad()
            
            train_loss += loss.item() * CONFIG["accumulate_grad_batches"]
            
            # Store predictions
            with torch.no_grad():
                preds = torch.argmax(outputs, dim=1)
                train_preds.extend(preds.cpu().numpy())
                train_labels.extend(labels.cpu().numpy())
            
            pbar.set_postfix({'loss': f'{loss.item():.4f}'})
        
        # Calculate training metrics
        train_loss /= len(train_loader)
        train_acc = accuracy_score(train_labels, train_preds)
        train_precision = precision_score(train_labels, train_preds, average='weighted', zero_division=0)
        train_recall = recall_score(train_labels, train_preds, average='weighted', zero_division=0)
        train_f1 = f1_score(train_labels, train_preds, average='weighted', zero_division=0)
        
        # Validation phase
        val_loss, val_preds, val_labels = evaluate_model(model, val_loader, criterion)
        val_acc = accuracy_score(val_labels, val_preds)
        val_precision = precision_score(val_labels, val_preds, average='weighted', zero_division=0)
        val_recall = recall_score(val_labels, val_preds, average='weighted', zero_division=0)
        val_f1 = f1_score(val_labels, val_preds, average='weighted', zero_division=0)
        
        # Update scheduler
        scheduler.step(val_acc)
        
        # Store history
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        history['train_precision'].append(train_precision)
        history['val_precision'].append(val_precision)
        history['train_recall'].append(train_recall)
        history['val_recall'].append(val_recall)
        history['train_f1'].append(train_f1)
        history['val_f1'].append(val_f1)
        history['learning_rates'].append(optimizer.param_groups[0]['lr'])
        
        # Early stopping
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            patience_counter = 0
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_acc': val_acc,
                'history': history
            }, 'best_model.pth')
        else:
            patience_counter += 1
        
        print(f"Epoch {epoch+1:3d} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
        print(f"         | Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f} | Train F1: {train_f1:.4f} | Val F1: {val_f1:.4f}")
        print(f"         | LR: {optimizer.param_groups[0]['lr']:.2e} | Best Val Acc: {best_val_acc:.4f}")
        
        if patience_counter >= CONFIG["early_stopping_patience"]:
            print(f"Early stopping at epoch {epoch+1}")
            break
    
    # Load best model
    checkpoint = torch.load('best_model.pth')
    model.load_state_dict(checkpoint['model_state_dict'])
    
    return history, val_preds, val_labels

def evaluate_model(model, loader, criterion):
    model.eval()
    total_loss = 0.0
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(loader, desc="Evaluating", leave=False):
            inputs, labels = inputs.to(device), labels.to(device)
            
            with torch.cuda.amp.autocast():
                outputs = model(inputs)
                loss = criterion(outputs, labels)
            
            total_loss += loss.item()
            preds = torch.argmax(outputs, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    return total_loss / len(loader), all_preds, all_labels

def test_time_augmentation(model, loader, tta_transforms):
    model.eval()
    all_preds = []
    all_labels = []
    all_probs = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(loader, desc="TTA Inference"):
            inputs, labels = inputs.to(device), labels.to(device)
            
            # Collect predictions from all augmentations
            tta_outputs = []
            for _ in range(2):  # Reduced TTA iterations
                with torch.cuda.amp.autocast():
                    outputs = model(inputs)
                    tta_outputs.append(F.softmax(outputs, dim=1))
            
            # Average predictions
            avg_outputs = torch.stack(tta_outputs).mean(dim=0)
            preds = torch.argmax(avg_outputs, dim=1)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(avg_outputs.cpu().numpy())
    
    return np.array(all_preds), np.array(all_labels), np.array(all_probs)

# Comprehensive Visualization Function
def plot_comprehensive_results(history, class_names, val_preds, val_labels, test_preds=None, test_labels=None, test_probs=None):
    # Use default matplotlib style
    plt.style.use('default')
    fig = plt.figure(figsize=(20, 15))
    
    # Color schemes
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD']
    
    # 1. Training History - Loss
    plt.subplot(3, 4, 1)
    plt.plot(history['train_loss'], label='Training Loss', color=colors[0], linewidth=2)
    plt.plot(history['val_loss'], label='Validation Loss', color=colors[1], linewidth=2)
    plt.title('Training & Validation Loss', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 2. Training History - Accuracy
    plt.subplot(3, 4, 2)
    plt.plot(history['train_acc'], label='Training Accuracy', color=colors[0], linewidth=2)
    plt.plot(history['val_acc'], label='Validation Accuracy', color=colors[1], linewidth=2)
    plt.title('Training & Validation Accuracy', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 3. F1-Score Evolution
    plt.subplot(3, 4, 3)
    plt.plot(history['train_f1'], label='Training F1', color=colors[0], linewidth=2)
    plt.plot(history['val_f1'], label='Validation F1', color=colors[1], linewidth=2)
    plt.title('F1-Score Evolution', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch')
    plt.ylabel('F1-Score')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 4. Learning Rate Schedule
    plt.subplot(3, 4, 4)
    plt.plot(history['learning_rates'], color=colors[2], linewidth=2)
    plt.title('Learning Rate Schedule', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.yscale('log')
    plt.grid(True, alpha=0.3)
    
    # 5. Validation Confusion Matrix
    plt.subplot(3, 4, 5)
    cm_val = confusion_matrix(val_labels, val_preds)
    sns.heatmap(cm_val, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Validation Confusion Matrix', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    
    # 6. Test Confusion Matrix
    if test_preds is not None and test_labels is not None:
        plt.subplot(3, 4, 6)
        cm_test = confusion_matrix(test_labels, test_preds)
        sns.heatmap(cm_test, annot=True, fmt='d', cmap='Greens', 
                    xticklabels=class_names, yticklabels=class_names)
        plt.title('Test Confusion Matrix', fontsize=14, fontweight='bold')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
    
    # 7. ROC Curves
    if test_probs is not None and test_labels is not None:
        plt.subplot(3, 4, 7)
        n_classes = len(class_names)
        test_labels_bin = label_binarize(test_labels, classes=range(n_classes))
        
        for i in range(n_classes):
            fpr, tpr, _ = roc_curve(test_labels_bin[:, i], test_probs[:, i])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, color=colors[i % len(colors)], lw=2,
                    label=f'{class_names[i]} (AUC = {roc_auc:.3f})')
        
        plt.plot([0, 1], [0, 1], 'k--', lw=2)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curves', fontsize=14, fontweight='bold')
        plt.legend()
        plt.grid(True, alpha=0.3)
    
    # 8. Class-wise Performance Metrics
    plt.subplot(3, 4, 8)
    if test_preds is not None and test_labels is not None:
        test_precision = precision_score(test_labels, test_preds, average=None)
        test_recall = recall_score(test_labels, test_preds, average=None)
        test_f1 = f1_score(test_labels, test_preds, average=None)
        
        x = np.arange(len(class_names))
        width = 0.25
        
        plt.bar(x - width, test_precision, width, label='Precision', color=colors[0], alpha=0.8)
        plt.bar(x, test_recall, width, label='Recall', color=colors[1], alpha=0.8)
        plt.bar(x + width, test_f1, width, label='F1-Score', color=colors[2], alpha=0.8)
        
        plt.xlabel('Classes')
        plt.ylabel('Score')
        plt.title('Class-wise Performance Metrics', fontsize=14, fontweight='bold')
        plt.xticks(x, class_names, rotation=45)
        plt.legend()
        plt.grid(True, alpha=0.3)
    
    # 9. Precision and Recall Evolution
    plt.subplot(3, 4, 9)
    plt.plot(history['train_precision'], label='Training Precision', color=colors[0], linewidth=2)
    plt.plot(history['val_precision'], label='Validation Precision', color=colors[1], linewidth=2)
    plt.plot(history['train_recall'], label='Training Recall', color=colors[2], linewidth=2, linestyle='--')
    plt.plot(history['val_recall'], label='Validation Recall', color=colors[3], linewidth=2, linestyle='--')
    plt.title('Precision & Recall Evolution', fontsize=14, fontweight='bold')
    plt.xlabel('Epoch')
    plt.ylabel('Score')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 10. Model Performance Comparison
    plt.subplot(3, 4, 10)
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
    if test_preds is not None and test_labels is not None:
        val_metrics = [
            accuracy_score(val_labels, val_preds),
            precision_score(val_labels, val_preds, average='weighted'),
            recall_score(val_labels, val_preds, average='weighted'),
            f1_score(val_labels, val_preds, average='weighted')
        ]
        test_metrics = [
            accuracy_score(test_labels, test_preds),
            precision_score(test_labels, test_preds, average='weighted'),
            recall_score(test_labels, test_preds, average='weighted'),
            f1_score(test_labels, test_preds, average='weighted')
        ]
        
        x = np.arange(len(metrics))
        width = 0.35
        
        plt.bar(x - width/2, val_metrics, width, label='Validation', color=colors[0], alpha=0.8)
        plt.bar(x + width/2, test_metrics, width, label='Test', color=colors[1], alpha=0.8)
        
        plt.xlabel('Metrics')
        plt.ylabel('Score')
        plt.title('Validation vs Test Performance', fontsize=14, fontweight='bold')
        plt.xticks(x, metrics, rotation=45)
        plt.legend()
        plt.grid(True, alpha=0.3)
    
    # 11. Final Performance Summary
    plt.subplot(3, 4, 11)
    final_metrics = {
        'Final Train Acc': history['train_acc'][-1],
        'Final Val Acc': history['val_acc'][-1],
        'Best Val Acc': max(history['val_acc']),
        'Final Train F1': history['train_f1'][-1],
        'Final Val F1': history['val_f1'][-1]
    }
    
    if test_preds is not None and test_labels is not None:
        final_metrics['Test Acc'] = accuracy_score(test_labels, test_preds)
        final_metrics['Test F1'] = f1_score(test_labels, test_preds, average='weighted')
    
    plt.barh(range(len(final_metrics)), list(final_metrics.values()), color=colors[:len(final_metrics)])
    plt.yticks(range(len(final_metrics)), list(final_metrics.keys()))
    plt.xlabel('Score')
    plt.title('Final Performance Summary', fontsize=14, fontweight='bold')
    plt.grid(True, alpha=0.3)
    
    # 12. Class Distribution
    plt.subplot(3, 4, 12)
    val_class_counts = Counter(val_labels)
    class_names_counts = [f"{class_names[i]}\n({val_class_counts.get(i, 0)})" for i in range(len(class_names))]
    plt.pie(val_class_counts.values(), labels=class_names_counts, autopct='%1.1f%%', colors=colors[:len(class_names)])
    plt.title('Validation Class Distribution', fontsize=14, fontweight='bold')
    
    plt.tight_layout()
    plt.show()

# Main execution function
def main():
    print("Starting Colon Cancer Classification Training...")
    
    # Prepare data
    train_loader, val_loader, test_loader, class_names, class_weights = prepare_data()
    
    # Initialize model
    model = EnhancedClassifier(num_classes=len(class_names), model_name=CONFIG["model_name"])
    model = model.to(device)
    
    print(f"\nModel initialized with {sum(p.numel() for p in model.parameters()):,} parameters")
    print(f"Trainable parameters: {sum(p.numel() for p in model.parameters() if p.requires_grad):,}")
    
    # Train model
    print("\nStarting training...")
    history, val_preds, val_labels = train_model(model, train_loader, val_loader, class_weights)
    
    # Test evaluation
    print("\nEvaluating on test set...")
    if CONFIG["test_time_augmentation"]:
        test_preds, test_labels, test_probs = test_time_augmentation(model, test_loader, get_transforms()[2])
    else:
        criterion = nn.CrossEntropyLoss()
        _, test_preds, test_labels = evaluate_model(model, test_loader, criterion)
        test_probs = None
    
    # Print final results
    print("\n" + "="*50)
    print("FINAL RESULTS")
    print("="*50)
    
    val_acc = accuracy_score(val_labels, val_preds)
    val_f1 = f1_score(val_labels, val_preds, average='weighted')
    test_acc = accuracy_score(test_labels, test_preds)
    test_f1 = f1_score(test_labels, test_preds, average='weighted')
    
    print(f"Validation Accuracy: {val_acc:.4f}")
    print(f"Validation F1-Score: {val_f1:.4f}")
    print(f"Test Accuracy: {test_acc:.4f}")
    print(f"Test F1-Score: {test_f1:.4f}")
    
    # Detailed classification report
    print("\nDetailed Classification Report (Test Set):")
    print(classification_report(test_labels, test_preds, target_names=class_names))
    
    # Visualize results
    plot_comprehensive_results(history, class_names, val_preds, val_labels, test_preds, test_labels, test_probs)
    
    # Save final model
    torch.save({
        'model_state_dict': model.state_dict(),
        'class_names': class_names,
        'config': CONFIG,
        'test_accuracy': test_acc,
        'test_f1': test_f1
    }, 'final_model.pth')
    
    print("\nTraining completed! Final model saved as 'final_model.pth'")

if __name__ == "__main__":
    main()
```
