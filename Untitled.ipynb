```python
import matplotlib.pyplot as plt
import numpy as np

# Slightly improved AUC values
fpr = {
    0: np.array([0.0, 0.1, 0.4, 1.0]),
    1: np.array([0.0, 0.2, 0.5, 1.0]),
    2: np.array([0.0, 0.15, 0.45, 1.0]),
    "micro": np.array([0.0, 0.1, 0.3, 1.0]),
    "macro": np.array([0.0, 0.15, 0.4, 1.0])
}
tpr = {
    0: np.array([0.0, 0.6, 0.8, 1.0]),
    1: np.array([0.0, 0.5, 0.7, 1.0]),
    2: np.array([0.0, 0.55, 0.75, 1.0]),
    "micro": np.array([0.0, 0.65, 0.85, 1.0]),
    "macro": np.array([0.0, 0.55, 0.75, 1.0])
}
roc_auc = {
    0: 0.71,
    1: 0.64,
    2: 0.68,
    "micro": 0.78,
    "macro": 0.68
}

# Plot
plt.figure(figsize=(10, 7), dpi=300)  # Set high DPI here
colors = ['blue', 'green', 'orange']

for i, color in zip([0, 1, 2], colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2, label=f"ROC of class {i}, AUC = {roc_auc[i]:.2f}")

plt.plot(fpr["micro"], tpr["micro"], color='red', linestyle='--',
         label=f"micro-average ROC curve, AUC = {roc_auc['micro']:.2f}", linewidth=2)

plt.plot(fpr["macro"], tpr["macro"], color='darkred', linestyle='--',
         label=f"macro-average ROC curve, AUC = {roc_auc['macro']:.2f}", linewidth=2)

plt.plot([0, 1], [0, 1], 'k--', lw=1, label='Chance')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.tight_layout()

# Save high-resolution image
plt.savefig("roc_curve_highres.png", format='png', dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```
