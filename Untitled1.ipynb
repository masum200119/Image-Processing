```python
import matplotlib.pyplot as plt

# Data from the image
locations = [
    "Time of accident",
    "Vehicular Involvement",
    "Hatikumrul Pabna",
    "Pabna Natore Triangle",
    "Ullapara Pabna",
    "Tarash Thana Highway",
    "Baghabari-Siraiganj Hifhway",
    "Pabna Rajapur Highway"
]

importance_values = [0.7, 0.55, 0.33, 0.3, 0.26, 0.21, 0.12, 0.1]

# Create the bar plot
plt.figure(figsize=(10, 6))
bars = plt.barh(locations, importance_values, color='skyblue')

# Customize the plot (without showing values)
plt.xlabel('Variable Importance', fontsize=12)
# plt.title('Feature Importance of Accident Locations', fontsize=14)
plt.gca().invert_yaxis()  # Highest importance at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Save the plot with high DPI
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=300, bbox_inches='tight')  # High-resolution image
plt.show()
```
